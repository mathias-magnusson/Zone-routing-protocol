```python
import pandas as pd

filepath_1 = "/Users/mathiasmagnusson/Zone-routing-protocol/iarp-test-1.csv"
filepath_2 = "/Users/mathiasmagnusson/Zone-routing-protocol/iarp-test-2.csv"
filepath_3 = "/Users/mathiasmagnusson/Zone-routing-protocol/iarp-test-3.csv"
filepath_4 = "/Users/mathiasmagnusson/Zone-routing-protocol/iarp-test-4.csv"

df1 = pd.read_csv(filepath_1)
df2 = pd.read_csv(filepath_2)
df3 = pd.read_csv(filepath_3)
df4 = pd.read_csv(filepath_4)
df1 = df1.T
df2 = df2.T
df3 = df3.T
df4 = df4.T
df1.columns = ["$\u03C1$ = 1"]
df2.columns = ["$\u03C1$ = 2"]
df3.columns = ["$\u03C1$ = 3"]
df4.columns = ["$\u03C1$ = 4"]
```

```python
## IARP RUNNING TIMES OF p = 1,2,3,4,5

import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')

_, ax = plt.subplots(figsize=(20,6))
df1.plot(ax=ax, linewidth=3)
df2.plot(ax=ax, linewidth=3)
df3.plot(ax=ax, linewidth=3)
df4.plot(ax=ax, linewidth=3)

ax.set_yscale('log')

ax.vlines(range(0, 450, 100), 0, 100000, colors="tab:gray", label='Orbit period')
ax.set_ylabel("Simulation units", weight="bold", fontsize=14)
ax.set_xlabel("Timestamp (minutes)", weight="bold", fontsize=14)
plt.title("IARP running time for node 0", weight="bold", fontsize=18)
plt.ylim(0,100000)
plt.xlim(0,200)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

legend = plt.legend(loc='upper right', bbox_to_anchor=(0.98, 0.98), frameon=True, framealpha=1, facecolor='white', fontsize = 14)
```

```python
## IARP RUNNING TIME AT p = 2

plt.style.use('seaborn-v0_8')

_, ax = plt.subplots(figsize=(20,6))
df2.plot(ax=ax, linewidth=3)

ax.vlines(range(0, 450, 100), 0, 5000, colors="tab:gray", label='Orbit period')
ax.set_ylabel("Simulation units", weight="bold", fontsize=14)
ax.set_xlabel("Timestamp (minutes)", weight="bold", fontsize=14)
plt.title("IARP running time for node 0", weight="bold", fontsize=18)
plt.ylim(0,200)
plt.xlim(0,300)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)


#ax.legend(loc='upper right')
legend = plt.legend(loc='upper right', bbox_to_anchor=(0.98, 0.98), frameon=True, framealpha=1, facecolor='white', fontsize = 14)
```

```python
## IARP RUNNING TIME AS A FUNCTION OF ZONE RADII
import seaborn as sns

running_times_min = [2.39, 22.19, 245.80, 3474, 50499.8]
running_times_max = [6.79, 145.6, 2386.80, 33821.19, 431215]
p_values = [1, 2, 3, 4, 5]

plt.style.use('seaborn-v0_8')

plt.plot(p_values, running_times_min, marker='o', linestyle='-', color='tab:blue', label='min')
plt.plot(p_values, running_times_max, marker='o', linestyle='-', color='tab:red', label='max')

plt.yscale('log')
plt.xticks([1, 2, 3, 4, 5], fontsize=12)
plt.yticks(fontsize=12)

plt.xlabel('$\u03C1$', weight='bold', fontsize=12)
plt.ylabel('Simulation units', weight='bold', fontsize=12)
plt.title('Minimum and maximum running times for IARP', weight='bold')
legend = plt.legend(loc='lower right', bbox_to_anchor=(0.98, 0.02), frameon=True, framealpha=1, facecolor='white', fontsize = 14)
```
