```python
import pandas as pd

filepath_1 = "/Users/mathiasmagnusson/Zone-routing-protocol/Data-out/iarp-test-1.csv"
filepath_2 = "/Users/mathiasmagnusson/Zone-routing-protocol/Data-out/iarp-test-2.csv"
filepath_3 = "/Users/mathiasmagnusson/Zone-routing-protocol/Data-out/iarp-test-3.csv"
filepath_4 = "/Users/mathiasmagnusson/Zone-routing-protocol/Data-out/iarp-test-only-0.csv"

df1 = pd.read_csv(filepath_1)
df2 = pd.read_csv(filepath_2)
df3 = pd.read_csv(filepath_3)
df4 = pd.read_csv(filepath_4)
df1 = df1.T
df2 = df2.T
df3 = df3.T
df4 = df4.T
df1.columns = ["$\u03C1$ = 1"]
df2.columns = ["$\u03C1$ = 2"]
df3.columns = ["$\u03C1$ = 3"]
df4.columns = ["$\u03C1$ = 2"]
```

```python
## IARP RUNNING TIMES OF p = 1,2,3,4,5

plt.style.use('seaborn-v0_8')

_, ax = plt.subplots(figsize=(20,6))
df1.plot(ax=ax, linewidth=4)
df2.plot(ax=ax, linewidth=4)
df3.plot(ax=ax, linewidth=3)
#df4.plot(ax=ax, linewidth=3)

ax.set_yscale('log')

#ax.vlines(range(0, 450, 100), 0, 10000, colors="tab:gray", label='Orbit period')
ax.set_ylabel("# Packets", weight="bold", fontsize=20)
ax.set_xlabel("Time stamp (minutes)", weight="bold", fontsize=20)
plt.title("IARP packets for all nodes", weight="bold", fontsize=18)
plt.ylim(500,150000)
plt.xlim(0,100)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

legend = plt.legend(loc='upper right', bbox_to_anchor=(0.98, 0.98), frameon=True, framealpha=1, facecolor='white', fontsize = 14)
```

```python
## IARP RUNNING TIME AT p = 2

plt.style.use('seaborn-v0_8')

_, ax = plt.subplots(figsize=(20,6))
df4.plot(ax=ax, linewidth=3)

ax.vlines(range(0, 450, 100), 0, 300, colors="tab:gray", label='Orbit period')
ax.set_ylabel("# Packets", weight="bold", fontsize=20)
ax.set_xlabel("Time stamp (minutes)", weight="bold", fontsize=20)
plt.title("IARP packets for node 0", weight="bold", fontsize=18)
plt.ylim(0,300)
plt.xlim(0,200)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

legend = plt.legend(loc='upper right', bbox_to_anchor=(0.98, 0.98), frameon=True, framealpha=1, facecolor='white', fontsize = 14)
```
