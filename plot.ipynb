```python
import pandas as pd

filepath_1 = "/Users/mathiasmagnusson/Zone-routing-protocol/iarp-test-1.csv"
filepath_2 = "/Users/mathiasmagnusson/Zone-routing-protocol/iarp-test-2.csv"
filepath_3 = "/Users/mathiasmagnusson/Zone-routing-protocol/iarp-test-3.csv"

df1 = pd.read_csv(filepath_1)
df2 = pd.read_csv(filepath_2)
df3 = pd.read_csv(filepath_3)
df1 = df1.T
df2 = df2.T
df3 = df3.T
df1.columns = ["$\u03C1$ = 1"]
df2.columns = ["$\u03C1$ = 2"]
df3.columns = ["$\u03C1$ = 3"]
```

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')

_, ax = plt.subplots(figsize=(20,6))
df1.plot(ax=ax)
df2.plot(ax=ax)
df3.plot(ax=ax)

ax.set_yscale('log')

ax.vlines(range(0, 450, 100), 0, 5000, colors="tab:orange", label='Orbit periods')
ax.set_ylabel("Simulation units")
ax.set_xlabel("Timestamp (minutes)")
plt.title("IARP running time for node 0", weight="bold", fontsize=18)
plt.ylim(0,3000)
plt.xlim(0,200)
ax.legend(loc='upper right')

```
